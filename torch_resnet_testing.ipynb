# ResNet18v1 validate<a id="01"></a>

```python
resnet18 = torch_resnet.ResNet18(v=1.0)
summary(resnet18, input_size=(1, 3, 224, 224), depth=4)
```

```python
from torchview import draw_graph
model_graph = draw_graph(resnet18, input_size=(1, 3, 224, 224), expand_nested=True, depth=4)

model_graph.visual_graph
```

```python
torch_renet18 = models.resnet18(weights=None)
summary(torch_renet18, input_size=(1, 3, 224, 224), depth=4)
```
